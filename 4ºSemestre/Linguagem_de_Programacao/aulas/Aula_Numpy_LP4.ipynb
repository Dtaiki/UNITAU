## Data Type Objects (dtype)
O método data type descreve a interpretação do bloco de memória fixo correspondendo a um array, dependendo dos seguintes aspectos:


*   tipo do dado (inteiro, float ou objeto Python)
*   tamanho do dado


*  Byte order (little-endian ou big-endian)
*  No caso de tipo estruturado, os nomes dos campos, tipo do dado de cada campo e parte do bloco da memória pega por cada bloco.


*   Se o tipo do dado é subarray, seu formato e tipo do dado

```python
np.dtype(object, align, copy)
```

os parametros são:


*   **Object**: a ser convertido para o tipo de dado object
*   **Align**: Se True, addiciona preenchimento ao campo para torná-lo similar a estrutruta


*   **Copy**: Cria nova cópia do objeto

Exemplo 1: array do tipo escalar

```python
dt = np.dtype(np.int32)
print(dt)
```

Exemplo 2: int8, int16, int32, int64 pode ser substituido pela string equivalnte 'i1', 'i2', 'i4', etc.

```python
dt = np.dtype('i4')
print(dt)
```

Exemplo 4: criando um tipo de dado estruturado

```python
dt = np.dtype([('age', np.int8)])
#
# agora aplicando para um objeto ndarray
a = np.array([(10,), (20,), (30,)], dtype = dt)
print(a)
```

```python
# o nome definido 'age' pode ser usado para acessar o conteúdo
print(a['age'])
```

Exemplo 5: um dado estruturado chamado **student** com campo do tipo string 'name' e um c**ampo inteiro** 'age' e outro **float** 'marks'.

```python
student = np.dtype([('name', 'S20'), ('age', 'i1'), ('marks', 'f4')])
print(student)
```

```python
a = np.array([('a', 21, 50), ('xyz', 18, 75)], dtype=student)
print(a)
#
print(a['name'])
```

# NumPy - Atributos

## **ndarray.shape**

Este método retorna uma tupla contendo a dimensão do array. Também pode ser usado para redimensionar o array

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(type(a))
print(a,'\n','shape = ', a.shape)
```

```python
a.shape = (3,2)
print(a)
```

Numpy também fornece uma função reshape para redimensionar.

```python
b = a.reshape(3,2)
print(b)
```

**ndarray.ndim**
Este método retorna o numero da dimensão do array

```python
a = np.arange(24)
print(a)
print('dim a = ', a.ndim, '\n')
#
# redimensionando
b = a.reshape(2,4,3)
print(b)
```

**numpy.itemsize**
Método que retorna o tamanho de cada elemento do array in bytes

```python
x = np.array([1,2,3,4,5], dtype=np.int8)
print(x.itemsize)
```

```python
x = np.array([1,2,3,4,5], dtype=np.float32)
print(x.itemsize)
```

**numpy.flags**
O objeto ndarray possui os seguintes atributos:


*   **C_CONTIGUOUS(C)**: O dado é single, segmento contínuo estilo C.
*   **F_CONTIGUOUS(F)**: O dado é single, segmento contínuo estiLo Fortran.
*   **OWNDATA(O)**:o array possui a mémoria que usa
*   **WRITEABLE(W)**: a area do dado pode ser escrita também.
*   **ALIGNED(A)**: o dado e todos seus elementos estão alinhados com o hardware.
*   **UPDATEIFCOPY(U)**: este array é copia de alguma outra.


```python
print(x.flags)
```

# NumPy - rotinas para criação de arrays

**numpy.empty**

Cria um array não inicializada no formato de tipo de dado especificado:

o construtor tem os seguintes parametros:
**shape**: formato do array em ***int*** ou tupla de ***int***
**dtype**: tipo de dado desejado. opcional.
**order**: 'C' para estilo-C, 'F' para estilo Fortran.

**numpy.empty**(shape, dtype=float, order='C')

```python
xempty = np.empty([3,2], dtype = np.float64)
print(xempty)
```

# **numpy.zeros**

Cria um array de zeros, no formato de tipo de dado especificado:
**numpy.zeros**(shape, dtype, order='C')

```python
x = np.zeros([5,4])
print(x)
```

# **numpy.ones**

Cria um array de valores unitários, no formato de tipo de dado especificado: numpy.zeros(shape, dtype, order='C')

```python
x = np.ones([5,2],dtype = np.int32)
print(x)
```

# NumPy - Arrays criadas por meio de dados existentes

# **numpy.asarray**

**numpy.asarray**(a, dtype=None, order=None)

Similar ao numpy.array, mas com menos parametros.

**a**: dado de entrada que pode ser uma lista, lista de tuplas, tuplas de tuplas ou tuplas de listas.

```python
x = [1, 2, 3]
print(type(x))
a = np.asarray(x)
print(type(a), a)
#
#
b = np.asarray(x, dtype = float)
print(b)
```

# **numpy.fromiter**
**numpy.fromiter**(iterable, dtype, count=-1)

**iterable**: qualquer objeto iterável

**count**: o numero de ítens para ser lido vindo do iterador. O valor padrão é -1, que implica na leitura de todo o dado.

```python
lista = range(5)
it = iter(lista)
x = np.fromiter(it, dtype=float)
print(x)
```

# Numpy - Arrays vindo de Ranges

# **numpy.arange**

**numpy.arange**(start, stop, step, dtype)

```python
x = np.arange(5)
print(x)
#
#
x = np.arange(5,dtype=float)
print(x)
```

# **numpy.linspace**

**numpy.linspace**(start, stop, num, endpoint, restep, dtype)

```python
x = np.linspace(0, 50, 51)
print(x)
```

```python
x = np.linspace(0, 30, 16)
print(x)
```

```python
x = np.linspace(0, 30, 16, endpoint=False)
print(x)
```

```python
x,step = np.linspace(0,30,16, endpoint = False, retstep=True)
print(x)
print("Step = ", step)
```
